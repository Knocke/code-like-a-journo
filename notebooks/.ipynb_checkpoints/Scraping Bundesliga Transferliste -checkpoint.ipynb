```python
import requests 
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
r = requests.get("http://embed.scribblelive.com/Embed/v7.aspx?Id=2592084&ThemeId=36126")
soup = BeautifulSoup(r.text, 'lxml')
```

```python
soup
```

```python
verein = []
data = []
zugänge = []
abgänge = []
```

```python
for h2 in soup.find_all('h2'):
    verein.append(h2.text)
```

```python
verein
```

```python
for p in soup.find_all('p'):
    data.append(p.text)
```

```python
for x in data: 
    if re.search(r'Zugänge:', x):
        zugänge.append(x)
```

```python
for y in data:
    if re.search(r'Abgänge:', y):
        abgänge.append(y)
```

```python
columns = {'verein': verein, 'zugänge': zugänge, 'abgänge': abgänge}
```

```python
df = pd.DataFrame(columns)
```

```python
df['abgänge'] = df['abgänge'].str.split('Abgänge: ').str.get(1)
```

```python
df = df['verein', 'abgänge', 'zugänge']
```
